# Thư viện

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

# Import Data

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/My Drive/df_final_encoded.xlsx'
df = pd.read_excel(file_path)
df
```

```python
selected_columns = ["dtb1", "dtb2", "dtb3", "dtb4", "drl1", "drl2", "drl3", "drl4", "diem_tt", "sotc_tichluy", "khoa_encoded", "hedt_encoded"]
features = df[selected_columns]
labels = df["xeploai_encoded"]
features
```

```python
# Kiểm tra và xử lý giá trị thiếu
features.fillna(features.mean(), inplace=True)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Chia dữ liệu thành tập huấn luyện và kiểm tra (80:20)
X_train, X_test, y_train, y_test = train_test_split(features_scaled, labels, test_size=0.2, random_state=42)
```

```python
# Huấn luyện mô hình SVM với kernel RBF
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
# Đánh giá hiệu quả của SVM
print("SVM (RBF Kernel) Classification Report:")
print(classification_report(y_test, svm_predictions))
print("Accuracy:", accuracy_score(y_test, svm_predictions))
```

```python
# Huấn luyện mô hình Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
# Đánh giá hiệu quả của Random Forest Classifier
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, rf_predictions))
print("Accuracy:", accuracy_score(y_test, rf_predictions))
```
